```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from PIL import Image
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
train_path = "FruitinAmazon/train/"
test_path = "FruitinAmazon/test/"
```

```python
# List of all the files present in the train_path directory
os.listdir(train_path)
```

```python
# visualize one random image from each class directory 
def visualise(train_path):

  class_dirs = []
  for folder in os.listdir(train_path): # list all sub folder(representing classes) are added in the class_dirs list
    class_dirs.append(folder)

# Initialize empty list to store sample images and their corresponding classname
  images = []
  labels = []

  for class_dir in class_dirs:
    class_path = os.path.join(train_path, class_dir)

    image_files = []
    for image in os.listdir(class_path):
      image_files.append(image)

    random_image_file = random.choice(image_files) # Retrieve a random image from the folder
    image_path = os.path.join(class_path, random_image_file)

    images.append(plt.imread(image_path)) # append the image content to the images
    labels.append(class_dir)
      
# Set up a grid of subplots (2 rows with number of columns depending on the number of classes)
  fig, axes = plt.subplots(2, len(class_dirs) // 2, figsize = (12, 6))

    # Iterate over image and labels to determine where each image should be placed in the grid
  for i, (image, label) in enumerate(zip(images, labels)):
    row = i // (len(class_dirs) // 2)  
    col = i % (len(class_dirs) // 2)

    axes[row, col].imshow(image)
    axes[row, col].set_title(label)
    axes[row, col].axis('off')

  plt.show()
```

```python

visualise(train_path)
```

```python
def check_for_corrupted(dir):
    #Initialize subdirectories
  sub_dir = []
  for folder in os.listdir(dir):
    sub_dir.append(folder)

    #Iterate through subdirectories
  for dir in sub_dir:
    class_path = os.path.join(train_path, dir)

      #Collect image files
    image_files = []
    for image in os.listdir(class_path):
      image_files.append(image)

      for image in image_files:
        image = os.path.join(class_path, image)

        try:
          with Image.open(image) as img:
            img.verify()
        except Exception as e:
          print(f"Corrupted image found {image}")
  print("No corrupted image found")

check_for_corrupted(train_path)
```

Task 2: Loading and Preprocessing Image Data in keras:

```python
# preparing 3 dataset by splitting (train,test,validation)
# Make sure all dataset are resized and normalized
img_height = 128
img_width = 128
batch_size = 32
validation_split = 0.2

# Rescalling for normalization
rescale = tf.keras.layers.Rescaling(1./255)

# loading the training dataset (images)
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_path,
    labels='inferred',
    label_mode='int',
    image_size=(img_height, img_width),
    interpolation='nearest',
    batch_size=batch_size,
    shuffle=True,
    validation_split=validation_split,
    subset='training',
    seed=123
)


train_ds = train_ds.map(lambda x, y: (rescale(x), y))

# loading the validation dataset (images) to monitor models performance
val_ds = tf.keras.utils.image_dataset_from_directory(
    test_path,
    labels='inferred',
    label_mode='int',
    image_size=(img_height, img_width),
    interpolation='nearest',
    batch_size=batch_size,
    shuffle=False,
    seed=123
)

val_ds = val_ds.map(lambda x, y: (rescale(x), y))

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_path,
    labels='inferred',
    label_mode='int',
    image_size=(img_height, img_width),
    interpolation='nearest',
    batch_size=batch_size, # for faster computation
    shuffle=False,
    seed=123
)

test_ds = test_ds.map(lambda x, y: (rescale(x), y))
```

Task 3 - Implement a CNN with Convolutional Architecture:

```python
# Built sequential model
model = keras.Sequential([
    #Adds a convolutional layer with 32 filters, each having a kernel size of (3, 3).
    layers.Conv2D(32, (3, 3), padding='same', strides=1, activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2), strides=2), #reduces the feature map size by half (e.g., 128*128 -> 64x64).
    layers.Conv2D(32, (3, 3), padding='same', strides=1, activation='relu'),
    layers.MaxPooling2D((2, 2), strides=2),
    layers.Flatten(),
    layers.Dense(64, activation='relu'), # 64 neurons
    layers.Dense(128, activation='relu'),
    layers.Dense(6, activation="softmax")
])
```

```python
model.summary()
```

Task 4: Compile the Model

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

Task 4: Train the Model

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

callbacks = [
    keras.callbacks.ModelCheckpoint("best_model.keras"),
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True),
]

history = model.fit(
    train_ds,
    epochs=250,
    batch_size=16,
    validation_data=val_ds,
    callbacks=callbacks
)
```

```python
import matplotlib.pyplot as plt

train_loss = history.history['loss']
val_loss = history.history['val_loss']

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, len(train_loss) + 1), train_loss, label='Training Loss', color='blue')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validatation Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, len(train_acc) + 1), train_acc, label='Training Accuracy', color='blue')
plt.plot(range(1, len(val_acc) + 1), val_acc, label='Validation Accuracy', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validatation Loss')
plt.legend()

plt.show()
```

Task 5: Evaluate the Model

```python

test_loss, test_acc = model.evaluate(test_ds)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python

model.save('bestest_model.h5')

bestest_model = tf.keras.models.load_model('bestest_model.h5')
```

Task 7: Predictions and Classification Report

```python
import numpy as np
from sklearn.metrics import classification_report

y_pred_prob = bestest_model.predict(test_ds)
y_pred = np.argmax(y_pred_prob, axis=1)

y_true = []
for images, labels in test_ds.unbatch():
  y_true.append(labels.numpy())

y_true = np.array(y_true)
print(y_true)
print(classification_report(y_true, y_pred))
```
